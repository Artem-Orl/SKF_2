```python
# функции для проверки и оценки

# для проверки категориальных признаков


def check_bin_cat(dataframe, column):
    print(f'Уникальные значения для {column}:{dataframe[column].unique()}')
    print('='*50)
    print(
        f'Количество уникальных значений для {column}:{dataframe[column].nunique()}')
    print('='*50)
    display('Соотношение в %:',
            dataframe[column].value_counts(normalize=True)*100)
    if dataframe[column].nunique() > 15:
        plt.figure(figsize=(30, 8))
    elif dataframe[column].nunique() in range(8, 16):
        plt.figure(figsize=(12, 7))
    else:
        plt.figure(figsize=(6, 7))
    sns.countplot(dataframe[column])
    plt.title(column)
    plt.show()


# для проверки числовых признаков
def check_nums(data, column, log=False):
    dataframe = data.copy()
    if log == True:
        print('Данные были прологарифмированы!')
        dataframe[column] = dataframe[column].apply(lambda x: np.log(x+1))
    else:
        print('Данные НЕ были прологарифмированы!')
    Q1 = dataframe[column].quantile(0.25)
    Q3 = dataframe[column].quantile(0.75)
    IQR = Q3-Q1
    plt.figure(figsize=(10, 10))
    sns.displot(dataframe[column])
    plt.title(column)
    plt.show()
    print(f'для колонки:{column}', f'\nГраницы выбросов:{[Q1-1.5*IQR, Q3+1.5*IQR]}',
          f'\nКоличество выбросов:{dataframe[(dataframe[column] < (Q1-1.5*IQR)) | (dataframe[column] > (Q3+1.5*IQR))][column].count()}, %:{(dataframe[(dataframe[column] < (Q1-1.5*IQR)) | (dataframe[column] > (Q3+1.5*IQR))][column].count())/dataframe[column].count()}')
    print(
        f'Минимальное значение:{dataframe[column].min()}, Максимальное значение:{dataframe[column].max()}, Среднее значение:{round(dataframe[column].mean(),2)} ')


# боксплоты
def boxplot(dataset, col):
    fig, ax = plt.subplots(figsize=(10, 10))
    sns.boxplot(x=col, y='price', data=dataset[dataset[col].isin(
        dataset[col].value_counts().index[:15])], ax=ax)
    ax.set_title('Boxplot для '+col)
    plt.show()

# ttest


def get_stat_dif2(dataset, column):
    cols = dataset.loc[:, column].value_counts().index[:15]
    combinations_all = list(combinations(cols, 2))
    for comb in combinations_all:
        if ttest_ind(dataset.loc[dataset.loc[:, column] == comb[0], 'price'],
                     dataset.loc[dataset.loc[:, column] == comb[1], 'price'], equal_var=False).pvalue \
                <= 0.05/len(combinations_all):
            print('Найдены важные статистические различия для '+column)
            break


def mape(y_true, y_pred):
    return np.mean(np.abs((y_pred-y_true)/y_true))
```

```python
# Наивная модель
predicts = []
for index, row in pd.DataFrame(data_test[['model_info', 'productionDate']]).iterrows():
    query = f"model_info == '{row[0]}' and productionDate == {row[1]}"
    predicts.append(data_train.query(query)['price'].median())

# заполним не найденные совпадения
predicts = pd.DataFrame(predicts)
predicts = predicts.fillna(predicts.median())

# округлим
predicts = (predicts // 1000) * 1000

# оцениваем точность
print(
    f"Точность наивной модели по метрике MAPE: {(mape(data_test['price'], predicts.values[:, 0]))*100:0.2f}%")
```

# Объединение данных
Все обработки завернем в функцию, чтобы все было сконсолидировано в одном месте

```python
train['sample'] = 1  # помечаем где у нас трейн
test['sample'] = 0  # помечаем где у нас тест
# в тесте у нас нет значения price, мы его должны предсказать, поэтому пока просто заполняем нулями
test['price'] = 0

data = test.append(train, sort=False).reset_index(drop=True)  # объединяем
print(train.shape, test.shape, data.shape)
```

```python
def preproc_data(input_dataset):
    output_dataset = input_dataset.copy()

    # Обработка колонок для получения нужного вида признаков
    output_dataset.dropna(axis=0, subset=['Владельцы'], inplace=True)
    output_dataset['Владельцы'] = output_dataset['Владельцы'].apply(
        lambda x: int(x[0]))

    output_dataset['bodyType'] = output_dataset['bodyType'].apply(
        lambda x: x.split(sep=' ')[0])

    output_dataset['engineDisplacement'] = output_dataset['engineDisplacement'].apply(
        lambda x: x.replace('undefined LTR', '0'))  # электромобили, поэтому 0
    output_dataset['engineDisplacement'] = output_dataset['engineDisplacement'].apply(
        lambda x: float(x.split(sep=' ')[0]))

    output_dataset['enginePower'] = output_dataset['enginePower'].apply(
        lambda x: float(x.split(sep=' ')[0]))

    return output_dataset


df_preproc = preproc_data(data)
df_preproc.sample(5)
```

## Дополнительная обработка данных для модели
удаление лишних колонок, преобразование данных

```python
# выбросы в num колонках и датах?
# с перекодировкой категориальных и бинарных признаков нужно будет поэксперементировать, так как применять даже cat boost в лоб для колонок (для bodyType (13 уникальных значений), color(16), model_info (118)),


def data_optimizer(input_dataset):
    output_dataset = input_dataset.copy()

    # удаление лишних колонок
    output_dataset.drop(['Владение', 'vehicleConfiguration',
                         'name', 'sell_id', 'modelDate'], axis=1, inplace=True)
    output_dataset.drop(['description'], axis=1, inplace=True)

    # Обработка данных для построения модели
    # Логарифмирование
    # mileage
    output_dataset['mileage'] = output_dataset['mileage'].apply(
        lambda x: np.log(x+1))

    # enginePower
    output_dataset['enginePower'] = output_dataset['enginePower'].apply(
        lambda x: np.log(x+1))

    # Кодируем категориальные признаки
    # get_dummies
    output_dataset = pd.get_dummies(output_dataset, columns=[
                                    'bodyType', 'brand', 'color', 'fuelType', 'model_info', 'vehicleTransmission', 'ПТС', 'Привод', 'Руль'])

    # нормализуем количественные признаки для cat boost не принципиально, но для нейросети да
    #extended_num_cols=['engineDisplacement', 'enginePower', 'mileage', 'productionDate']
    num_cols.append('productionDate')
    scaler = MinMaxScaler()
    for col in num_cols:
        scaler.fit(output_dataset[output_dataset['sample'] == 1][[col]])
        output_dataset[col] = scaler.transform(output_dataset[[col]])

    return output_dataset


optimized_df = data_optimizer(df_preproc)

display(optimized_df.sample(5))
```

### Модель с Feature engineering и подбором гиперпараметров

```python
# пробуем рассчитать транспортный налог
# для электромобилей его нет


def tax(row):
    if row['enginePower'] > 250:
        return row['enginePower']*15

    elif row['enginePower'] > 200 and row['enginePower'] <= 250:
        return row['enginePower']*7.5

    elif row['enginePower'] > 150 and row['enginePower'] <= 200:
        return row['enginePower']*5

    elif row['enginePower'] > 100 and row['enginePower'] <= 150:
        return row['enginePower']*3.5

    elif row['enginePower'] <= 100:
        return row['enginePower']*2.5


optimized_df['tax'] = df_preproc.apply(lambda row: tax(row), axis=1)

# сразу возьмем логарифм и отномрмируем
optimized_df['tax'] = optimized_df['tax'].apply(lambda x: np.log(x+1))

scaler = MinMaxScaler()
scaler.fit(optimized_df[optimized_df['sample'] == 1][['tax']])
optimized_df['tax'] = scaler.transform(optimized_df[['tax']])
```

```python
# Также прологарифмируем цену
optimized_df['price'] = optimized_df['price'].apply(
    lambda x: np.log(x) if x > 0 else 0)
```

```python
# Строим модель опять

# выделим тестовую часть
train_data = optimized_df.query('sample == 1').drop(['sample'], axis=1)
test_data = optimized_df.query('sample == 0').drop(['sample'], axis=1)

y = train_data.price.values
X = train_data.drop(['price'], axis=1)
X_sub = test_data.drop(['price'], axis=1)

# Разбиваем
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.15, shuffle=True, random_state=RANDOM_SEED)

# подбор гиперпараметров
# grid_model=CatBoostRegressor(iterations=500,
# random_seed=RANDOM_SEED,
# early_stopping_rounds=50,
# eval_metric='MAPE',
# custom_metric=['RMSE', 'MAE'])

#grid = {'learning_rate': [0.03, 0.1,0.01,0.02,0.05,0.5,0.2,0.3],'depth': [4, 6, 10],'l2_leaf_reg': [1, 3, 5, 7, 9]}
#grid_search_result = grid_model.grid_search(grid, X=X_train, y=y_train, plot=True)
```

Лучшие параметры: {'params': {'depth': 4, 'l2_leaf_reg': 1, 'learning_rate': 0.3}

```python
# Строим саму финальную модель
fin_tab_model = CatBoostRegressor(iterations=5000,
                                  depth=4,
                                  l2_leaf_reg=1,
                                  # его уменьшим, так как для скорости при подборе гиперпараметров мы брали только 500 итераций
                                  learning_rate=0.033,
                                  random_seed=RANDOM_SEED,
                                  early_stopping_rounds=100,
                                  eval_metric='MAPE',
                                  custom_metric=['RMSE', 'MAE'],
                                  )
fin_tab_model.fit(X_train, y_train,
                  eval_set=(X_test, y_test),
                  use_best_model=True
                  )
```

```python
def make_augmentations(images):
    print('применение аугментаций', end='')
    augmented_images = np.empty(images.shape)
    for i in range(images.shape[0]):
        if i % 200 == 0:
            print('.', end='')
        augment_dict = augmentation(image=images[i])
        augmented_image = augment_dict['image']
        augmented_images[i] = augmented_image
    print('')
    return augmented_images
```

```python
# Без проброса признака
# упаковываем в датасеты. Цену берем без логарифма


def process_image(image):
    return augmentation(image=image.numpy())['image']


def tokenize_(descriptions):
    return sequence.pad_sequences(tokenize.texts_to_sequences(descriptions), maxlen=MAX_SEQUENCE_LENGTH)


def tokenize_text(text):
    return tokenize_([text.numpy().decode('utf-8')])[0]


def tf_process_train_dataset_element(image, table_data, text, price):
    im_shape = image.shape
    [image, ] = tf.py_function(process_image, [image], [tf.uint8])
    image.set_shape(im_shape)
    [text, ] = tf.py_function(tokenize_text, [text], [tf.int32])
    return (image, table_data, text), price


def tf_process_val_dataset_element(image, table_data, text, price):
    [text, ] = tf.py_function(tokenize_text, [text], [tf.int32])
    return (image, table_data, text), price


train_dataset = tf.data.Dataset.from_tensor_slices((
    images_train, X_train, data.description.iloc[X_train.index], y_train_nn
)).map(tf_process_train_dataset_element)

test_dataset = tf.data.Dataset.from_tensor_slices((
    images_test, X_test, data.description.iloc[X_test.index], y_test_nn
)).map(tf_process_val_dataset_element)

y_sub = np.zeros(len(X_sub))
sub_dataset = tf.data.Dataset.from_tensor_slices((
    images_sub, X_sub, data.description.iloc[X_sub.index], y_sub
)).map(tf_process_val_dataset_element)


# проверяем, что нет ошибок (не будет выброшено исключение):
train_dataset.__iter__().__next__()
test_dataset.__iter__().__next__()
sub_dataset.__iter__().__next__()
```

```python
# Второй вариант
# LR sheduler

# зададим функцию


def scheduler(epoch, lr):
    if (epoch > 29) and (epoch <= 49):
        return lr-0.0001
    if (epoch > 49) and (epoch <= 80):
        return lr-0.00005
    else:
        return lr


# Присвоим значение, чтобы внести в callback
lr_sheduler = tf.keras.callbacks.LearningRateScheduler(scheduler)


# Тест того, что все верно
def get_lr_metric(optimizer):
    def lr(y_true, y_pred):
        return optimizer.lr
    return lr


# LR sheduler
optimizer = tf.keras.optimizers.Adam()
# Чтобы отследить, что метрика задана верно
lr_metric = get_lr_metric(optimizer)


model.compile(loss='MAPE', optimizer=optimizer, metrics=['MAPE', lr_metric])
model.optimizer.lr = 0.005

checkpoint = ModelCheckpoint('../working/best_model.hdf5',
                             monitor='val_MAPE', verbose=0, mode='min', save_best_only=True)
earlystop = EarlyStopping(
    monitor='val_MAPE', patience=PATIENCE, restore_best_weights=True,)

# LR sheduler
callbacks_list = [checkpoint, earlystop, lr_sheduler]
```
